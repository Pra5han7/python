```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating an array of points
x=np.linspace(-np.pi,np.pi,256)
```

```python
C,S=np.cos(x),np.sin(x)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#setting x lim
#setting x lim
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(C.min()*1.1,C.max()*1.1)

ax.set_xlabel('x')
ax.set_ylabel('sin/cos(x)')
ax.plot(x,S,label='sin(x)')
ax.plot(x,C,label='cos(x)')

fig.legend(loc=0)
```

__setting limits for a typical sine plot__

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

#setting ticks to get rid of extra points from y and x axis
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks([-1,0,1])

#setting axes labels
ax.set_xlabel('x')
ax.set_ylabel('sin/cos(x)')

#setting legend
ax.plot(x,S,label='sin(x)')
ax.plot(x,C,label='cos(x)')

fig.legend(loc=0)
```

Ticks are now properly placed but their label is not very explicit. We could guess that 3.142 is  but it would be better to make it explicit. When we set tick values, we can also provide a corresponding label in the second argument list. Note that we'll use latex to allow for nice rendering of the label.

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

#setting ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

#adjusting spines
#we will remove top and right spines and set left and bottom as req
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

#setting axes labels
ax.set_xlabel('x')
ax.set_ylabel('sin/cos(x)')

#setting legend
ax.plot(x,S,label='sin(x)')
ax.plot(x,C,label='cos(x)')

fig.legend(loc=0)
```

__decoding spines__

```python
ax = plt.gca() #get current axis
ax.spines.keys()
```

ax=fig.add_axes([0,0,1,1]) gives us above axes and spines holds the border of the axes 

we can remove top and right border by setting their color to none instead of black. as left and bottom has plot labels removing them is not a good practise

```python
ax=plt.gca()
ax.spines['left'].set_color('none')
```

so better way is to remove top and right

```python
ax=plt.gca()
ax.spines['top'].set_color('none') #color can be set to red,blue or any here we just want that spine to disappear
ax.spines['right'].set_color('none')
ax.plot(x,S)
```

now we need to adjust position of left and bottom spines

```python
ax=plt.gca()
ax.spines['top'].set_color('none') #color can be set to red,blue or any here we just want that spine to disappear
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom') #sets the ticks to bottom of x axis
ax.spines['bottom'].set_position(('a',0)) # position can be outwards,axes or data
ax.yaxis.set_ticks_position('left') #sets the ticks to left of y axis
ax.spines['left'].set_position(('a',0))
```
